```python
import numpy as np
```

```python
def read_input(infile):
    data = []
    movement = []
    read_map = True
    with open(infile, 'r') as inf:
        for line in inf.readlines():
            if line.strip() == '':
                read_map = False
                continue
            if read_map:
                data.append([-1 if c =='#' else 0 if c =='.' else 1 if c == 'O' else 2 for c in line.strip()])
            else:
                movement += [c for c in line.strip()]

    return np.array(data, dtype=int), movement

def read_input_wide(infile):
    data = []
    movement = []
    read_map = True
    with open(infile, 'r') as inf:
        for line in inf.readlines():
            if line.strip() == '':
                read_map = False
                continue
            if read_map:
                l = []
                for c in line.strip():
                    if c == '#':
                        l.append(-1)
                        l.append(-1)
                    elif c == '.':
                        l.append(0)
                        l.append(0)
                    elif c == 'O':
                        l.append(1)
                        l.append(2)
                    else:
                        l.append(3)
                        l.append(0)

                data.append(l)
            else:
                movement += [c for c in line.strip()]

    return np.array(data, dtype=int), movement
            
```

```python
def pprint(data):
    return
    for l in range(data.shape[0]):
        print(''.join(['#' if v == -1 else '.' if v == 0 else 'O' if v == 1 else '@' for v in data[l,]]))

def pprint_wide(data):
    for l in range(data.shape[0]):
        print(''.join(['#' if v == -1 else '.' if v == 0 else '[' if v == 1 else ']' if v == 2 else '@' for v in data[l,]]))

def move(data, movement):

    y, x = np.argwhere(data==2)[0]

    dmap = {'>': (0, 1), 'v': (1, 0), '<': (0, -1), '^': (-1, 0)}

    for m in movement:
        if len(np.argwhere(data==2)) > 1:
            return

        dy, dx = dmap[m]
        ny, nx = y+dy, x+dx

        if data[ny,nx] < 0:
            # Wall
            continue
        elif data[ny, nx] == 0:
            # Free
            data[y, x] = 0
            data[ny, nx] = 2
            y, x = ny, nx
            continue
        else:
            if m == '>':
                l = data[y, x:]
                next_wall = np.min(np.argwhere(l == -1))
                where_free = np.argwhere(l==0)
                if len(where_free) == 0:
                    continue
                next_free = np.min(where_free)
                if next_free < next_wall:
                    # Space till the next wall
                    l[1:next_free+1] = l[:next_free]
                    l[0] = 0
                    y = ny
                    x = nx
            elif  m == '<':
                l = data[y, :x+1]
                next_wall = np.max(np.argwhere(l == -1))
                where_free = np.argwhere(l==0)
                if len(where_free) == 0:
                    continue
                next_free = np.max(where_free)
                if next_free > next_wall:
                    # Space till the next wall
                    l[next_free:-1] = l[next_free+1:]
                    l[-1] = 0
                    y = ny
                    x = nx
            if m == 'v':
                l = data[y:, x]
                next_wall = np.min(np.argwhere(l == -1))
                where_free = np.argwhere(l==0)
                if len(where_free) == 0:
                    continue
                next_free = np.min(where_free)
                if next_free < next_wall:
                    # Space till the next wall
                    l[1:next_free+1] = l[:next_free]
                    l[0] = 0
                    y = ny
                    x = nx

            elif m == '^':
                l = data[:y+1, x]
                next_wall = np.max(np.argwhere(l == -1))
                where_free = np.argwhere(l==0)
                if len(where_free) == 0:
                    # print('Wall')
                    continue
                next_free = np.max(where_free)
                if next_free > next_wall:
                    # Space till the next wall
                    l[next_free:-1] = l[next_free+1:]
                    l[-1] = 0
                    y = ny
                    x = nx
    # pprint(data)

    s = 0
    for cy, cx in np.argwhere(data == 1):
        s += 100*cy + cx
    return int(s)

def move_wide(data, movement):

    y, x = np.argwhere(data==3)[0]

    dmap = {'>': (0, 1), 'v': (1, 0), '<': (0, -1), '^': (-1, 0)}

    for m in movement:
        if len(np.argwhere(data==3)) > 1:
            return

        dy, dx = dmap[m]
        ny, nx = y+dy, x+dx

        if data[ny,nx] < 0:
            # Wall
            continue
        elif data[ny, nx] == 0:
            # Free
            data[y, x] = 0
            data[ny, nx] = 3
            y, x = ny, nx
            continue
        else:
            if m == '>':
                l = data[y, x:]
                next_wall = np.min(np.argwhere(l == -1))
                where_free = np.argwhere(l==0)
                if len(where_free) == 0:
                    continue
                next_free = np.min(where_free)
                if next_free < next_wall:
                    # Space till the next wall
                    l[1:next_free+1] = l[:next_free]
                    l[0] = 0
                    y = ny
                    x = nx
            elif  m == '<':
                l = data[y, :x+1]
                next_wall = np.max(np.argwhere(l == -1))
                where_free = np.argwhere(l==0)
                if len(where_free) == 0:
                    continue
                next_free = np.max(where_free)
                if next_free > next_wall:
                    # Space till the next wall
                    l[next_free:-1] = l[next_free+1:]
                    l[-1] = 0
                    y = ny
                    x = nx
            elif m == 'v':

                to_move = [[(y, x)]]
                if data[ny, nx] == 1:
                    to_move.append([(ny, nx), (ny, nx+1)])
                else:
                    to_move.append([(ny, nx), (ny, nx-1)])
                wall = False
                for l in range(ny+1, data.shape[1]):
                    nextline = []
                    for _, cx in to_move[-1]:
                        if data[l, cx] < 0:
                            wall = True
                            break
                        elif data[l, cx] == 1:
                            if (l, cx) not in nextline:
                                nextline.append((l, cx))
                            if (l, cx+1) not in nextline:
                                nextline.append((l, cx+1))
                        elif data[l, cx] == 2:
                            if (l, cx) not in nextline:
                                nextline.append((l, cx))
                            if (l, cx-1) not in nextline:
                                nextline.append((l, cx-1))
                    if wall:
                        break
                    if len(nextline) == 0:
                        # Move
                        for line in reversed(to_move):
                            for ly, lx, in line:
                                data[ly+1, lx] = data[ly,lx]
                                data[ly,lx] = 0
                        break
                    else:
                        to_move.append(nextline)
                if wall:
                    continue
                y = ny
                x = nx
            if m == '^':
                to_move = [[(y, x)]]
                if data[ny, nx] == 1:
                    to_move.append([(ny, nx), (ny, nx+1)])
                else:
                    to_move.append([(ny, nx), (ny, nx-1)])
                wall = False
                for l in range(ny-1, -1, -1):
                    nextline = []
                    for _, cx in to_move[-1]:
                        if data[l, cx] < 0:
                            wall = True
                            break
                        elif data[l, cx] == 1:
                            if (l, cx) not in nextline:
                                nextline.append((l, cx))
                            if (l, cx+1) not in nextline:
                                nextline.append((l, cx+1))
                        elif data[l, cx] == 2:
                            if (l, cx) not in nextline:
                                nextline.append((l, cx))
                            if (l, cx-1) not in nextline:
                                nextline.append((l, cx-1))
                    if wall:
                        break
                    if len(nextline) == 0:
                        for line in reversed(to_move):
                            for ly, lx, in line:
                                data[ly-1, lx] = data[ly,lx]
                                data[ly,lx] = 0
                        break
                    else:
                        to_move.append(nextline)
                if wall:
                    continue
                y = ny
                x = nx

    # pprint_wide(data)
    s = 0
    for cy, cx in np.argwhere(data == 1):
        s += 100*cy + cx
    return int(s)
```
