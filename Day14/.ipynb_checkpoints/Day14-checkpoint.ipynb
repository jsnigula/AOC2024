```python
import re
import numpy as np
```

```python
def read_input(infile):
    robots = []
    with open(infile, 'r') as inf:
        for line in inf.readlines():
            # print(line)
            robots.append(tuple(int(v) for v in re.match(r'p=(\d+),(\d+) v=(-?\d+),(-?\d+)', line.strip()).groups()))
            # print(robots)

    return robots
```

```python
def safe_space(robots, steps, sx, sy):

    quadrants = [0, 0, 0, 0]
    for r in robots:
        x,y,px,py = r
        nx = (x + px*steps) % sx
        ny = (y + py*steps) % sy

        if nx < sx//2:
            if ny < sy//2:
                quadrants[0] += 1
            elif ny > sy//2:
                quadrants[2] += 1
        elif nx > sx//2:
            if ny < sy//2:
                quadrants[1] += 1
            elif ny > sy//2:
                quadrants[3] += 1
        # print(nx, ny, quadrants)

    # print(quadrants)
    return quadrants[0] * quadrants[1] * quadrants[2] * quadrants[3]

def find_tree(robots):
    s = 0
    plot = np.zeros((103,101))
    while True:
        plot.fill(0)
        for r in robots:
            x,y,px,py = r
            plot[(y + py*s) % 103, (x + px*s) % 101] += 1
        plotted = False
        if len(np.argwhere(plot > 1)) == 0:
            print(s)
            pltstr = ''
            for l in range(101):
                pltstr += ''.join(['#' if v == 1 else ' ' for v in plot[l,:]]) + '\n'
            return s, pltstr
        # for i in range(0, 80, 5):
        #     if plotted: 
        #         break
        #     for j in range(0, 80,5):
        #         if np.sum(plot[i:i+21,j:j+21])>100:
        #             pltstr = ''
        #             for l in range(101):
        #                 pltstr += ''.join(['#' if v == 1 else ' ' for v in plot[l,:]]) + '\n'
        #             return s, pltstr
        s+=1
```

```python
print('*******\nPuzzle1\n*******\n')

print('Test case\n---------\n')

res = safe_space(read_input('input14a.txt'), 100, 11, 7)

print(f'Safety factor is {res}')

assert res == 12

print('\nPuzzle case\n-----------\n')

res = safe_space(read_input('input14.txt'), 100, 101, 103)

print(f'Safety factor is {res}')

assert res == 225552000

print('\n*******\nPuzzle2\n*******\n')

print('Puzzle case\n-----------\n')

res, plt = find_tree(read_input('input14.txt'))

print(f'Tree found after {res} steps')

assert res == 7371

print('And here is the plot!')

print(plt)
```
