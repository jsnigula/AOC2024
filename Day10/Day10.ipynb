```python
import numpy as np
from collections import defaultdict
```

```python
def read_input(infile):
    data = []
    with open(infile, 'r') as inf:
        for line in inf.readlines():
            data.append([int(c) if c != '.' else -1 for c in line.strip()])

    return np.array(data)

def find_trails(data):

    trails = defaultdict(list)
    trailheads = defaultdict(list)
    trailnum = 0
    for i in range(9, 0, -1):
        for y, x in np.argwhere(data == i):
            if i == 9:
                # New trail
                trails[(int(y),int(x))].append(trailnum)
                trailnum += 1
            elif (y, x) not in trails:
                continue

            for dy, dx in [(0, 1), (1, 0), (0, -1), (-1, 0)]:
                ny = int(y+dy)
                nx = int(x+dx)
                if ny >= 0 and ny < data.shape[0] and nx >= 0 and nx < data.shape[1]:
                    if data[ny, nx] == i-1:
                        trails[(ny, nx)] += trails[(y, x)]
                        if i == 1:
                            trailheads[(ny, nx)] += trails[(y, x)]

    score = 0 
    rating = 0
    for v in trailheads.values():
        score += len(set(v))
        rating += len(v)

    return score, rating

                
# len(np.argwhere(read_input('input10.txt')==9))
```

```python
print('*******\nPuzzle1\n*******\n')

print('Test case\n---------\n')

res, _ = find_trails(read_input('input10a.txt'))

print(f'Score is {res}')

assert res == 36

print('\nPuzzle case\n-----------\n')

res, _ = find_trails(read_input('input10.txt'))

print(f'Score is {res}')

assert res == 459

print('\n*******\nPuzzle2\n*******\n')

print('Test case\n---------\n')

_, res = find_trails(read_input('input10a.txt'))

print(f'Rating is {res}')

assert res == 81

print('\nPuzzle case\n-----------\n')

_, res = find_trails(read_input('input10.txt'))

print(f'Rating is {res}')

assert res == 1034
```
