```python
import numpy as np
```

```python
def read_input(infile):
    map = []
    with open(infile, 'r') as inf:
        for line in inf.readlines():
            map.append([0 if c == '.' else 2 if c == '^' else 1 for c in line.strip()])
    return np.array(map)            
```

```python
def turn(dy, dx):
    if (dy, dx) == (-1, 0):
        return 0, 1
    if (dy, dx) == (0, 1):
        return 1, 0
    if (dy, dx) == (1, 0):
        return 0, -1
    if (dy, dx) == (0, -1):
        return -1, 0

def print_map(map, visited):

    for r in range(map.shape[1]):
        l = ''
        for c in range(map.shape[0]):
            if map[r, c] == 1:
                l+='#'
            else:
                if (r, c, -1, 0) in visited:
                    l+='^'
                elif (r, c, 0, 1) in visited:
                    l+='>'
                elif (r, c, 1, 0) in visited:
                    l+='v'
                elif (r, c, 0, -1) in visited:
                    l+='<'
                else:
                    l+='.'
        print(l)

class Loop(Exception):
    pass

def find_path(map):
    visited = []
    y, x = np.argwhere(map==2)[0]
    y, x = int(y), int(x)
    dy, dx = -1, 0
    n_retraced = 0
    while True:

        if (y, x, dy, dx) in visited:
            n_retraced += 1
        if n_retraced > 2:
            # Found a loop
            raise Loop()
        if dy == 0:
            if dx > 0:
                w = np.argwhere(map[y, x:] == 1)
                if len(w) == 0:
                    for i in range(x, map.shape[1]):
                        visited.append((y, int(i), dy, dx))
                    return visited
                for i in range(x, x+w[0][0]-1):
                    visited.append((y, int(i), dy, dx))
                x = x + int(w[0][0]-1)
                visited.append((y, x, dy, dx))
                dy, dx = turn(dy, dx)
                # print_map(map, visited, y, x, dy, dx)
            if dx < 0:
                w = np.argwhere(map[y, :x] == 1)
                if len(w) == 0:
                    for i in range(x-1, -1, -1):
                        visited.append((y, int(i), dy, dx))
                    return visited
                for i in range(x, w[-1][0]+1, -1):
                    visited.append((y, int(i), dy, dx))
                x = int(w[-1][0]+1)
                visited.append((y, x, dy, dx))
                dy, dx = turn(dy, dx)
                # print_map(map, visited, y, x, dy, dx)
        else:
            if dy > 0:
                w = np.argwhere(map[y:, x] == 1)
                if len(w) == 0:
                    for i in range(y, map.shape[0]):
                        visited.append((int(i), x, dy, dx))
                    return visited
                for i in range(y, y+w[0][0]-1):
                    visited.append((int(i), x, dy, dx))
                y = y + int(w[0][0]-1)
                visited.append((y, x, dy, dx))
                dy, dx = turn(dy, dx)
                # print_map(map, visited, y, x, dy, dx)
            if dy < 0:
                w = np.argwhere(map[:y, x] == 1)
                if len(w) == 0:
                    for i in range(y-1, -1, -1):
                        visited.append((int(i), x, dy, dx))
                    return visited
                for i in range(y, w[-1][0]+1, -1):
                    visited.append((int(i), x, dy, dx))
                y = int(w[-1][0]+1)
                visited.append((y, x, dy, dx))
                dy, dx = turn(dy, dx)
                # print_map(map, visited, y, x, dy, dx)

def find_loops(map):

    visited = find_path(map)
    loop_pos = []
    
    sy, sx = np.argwhere(map==2)[0]

    for y, x, dy, dx in visited:

        ny, nx = y + dy, x + dx
        if ny >= map.shape[0] or ny < 0 or nx >= map.shape[1] or nx < 0:
            # Outside map
            continue
        newmap = np.copy(map)
        newmap[ny, nx] = 1

        if (ny, nx) == (sy, sx):
            continue
        if map[ny, nx] == 1:
            continue
        # print('Put obstacle at', ny, nx)
        # Check for a loop
        try:
            find_path(newmap)
        except Loop:
            # print('Found loop')
            loop_pos.append((ny, nx))                
    return len(set(loop_pos))

def find_distinct_pos(map):

    visited = find_path(map)
    pos = []
    for y, x, _, _ in visited:
        pos.append((y, x))

    return len(set(pos))
    
```

```python
print('*******\nPuzzle1\n*******\n')

print('Test case\n---------\n')

res = find_distinct_pos(read_input('input06a.txt'))

print(f'Number of steps is {res}')

assert res == 41

print('\nPuzzle case\n-----------\n')

res = find_distinct_pos(read_input('input06.txt'))

print(f'Number of steps is {res}')

assert res == 5086

print('\n*******\nPuzzle2\n*******\n')

print('Test case\n---------\n')

res = find_loops(read_input('input06a.txt'))

print(f'Number of positions is {res}')

assert res == 6

print('\nPuzzle case\n-----------\n')

res = find_loops(read_input('input06.txt'))

print(f'Number of positions is {res}')

assert res == 1770
```
