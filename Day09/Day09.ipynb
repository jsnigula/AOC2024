```python
def read_input(infile):
    with open(infile, 'r') as inf:
        return [int(c) for c in inf.readline().strip()]
```

```python
def chk(start, end):
    start = max(0, start)
    # print('c', start, end, (end**2 - start**2 + (end-start))/2) 
    return (end**2 - start**2 + (end-start))/2

def update(diskmap):

    ridx = len(diskmap)-1
    pos = 0
    chksum = 0
    i = 0
    while i <= len(diskmap):

        if diskmap[i] == 0:
            break
        chksum += i//2 * chk(pos-1, pos+diskmap[i]-1)
        pos += diskmap[i]
        # Check the next free space
        i+=1
        if i>ridx:
            break
        while diskmap[i] > 0:
            needed = diskmap[i]
            avail = diskmap[ridx]
            if avail >= diskmap[i]:
                diskmap[ridx] -= needed
                diskmap[i] -= needed
                chksum += (ridx+1)//2 * chk(pos-1, pos+needed-1)
                pos += needed
            else:
                diskmap[ridx] -= avail
                diskmap[i] -= avail
                chksum += (ridx+1)//2 * chk(pos-1, pos+avail-1)
                pos += avail

            if diskmap[ridx] == 0:
                ridx -= 2
        i+=1
    return int(chksum)
        
def reorder_and_check(diskmap):
    checked = []
    filemap = []
    for i in range(len(diskmap)):
        fileid = -1
        if i % 2 == 0:
            fileid = i // 2
        filemap.append([fileid, diskmap[i]])
    s_idx = 1
    r_idx = len(filemap)-1

    # print(filemap)

    r_idx = len(filemap)-1
    while r_idx > 0:
        # print('Trying to move', r_idx, filemap[r_idx])
        fileid = filemap[r_idx][0]
        needed = filemap[r_idx][1]
        # print('Needs', needed)
        if fileid == -1 or fileid in checked:
            r_idx -= 1
            continue
        for s_idx in range(1, r_idx):
            if filemap[s_idx][0] == -1 and filemap[s_idx][1] >= needed:
                filemap[s_idx][1] -= needed
                # print('Inserting between', filemap[s_idx], filemap[s_idx+1])
                filemap.insert(s_idx, [filemap[r_idx][0], filemap[r_idx][1]])
                filemap[r_idx][1] += needed
                filemap[r_idx+1][1] -= needed
                filemap[r_idx+1][0] = -1
                checked.append(fileid)
                break
        r_idx -= 1
        # print('fmap now', filemap)
        
    # print(filemap)
    # Calculate the new checksum
    chksum = 0
    p = 0
    for b in filemap:
        # print(b, p)
        if b[0] != -1:
            # print(b[0] * ((b[1]*b[1] + b[1])/2 + b[1] * (p-1)))
            chksum += b[0] * ((b[1]*b[1] + b[1])/2 + b[1] * (p-1))
        p += b[1]
    return int(chksum)
```

```python
print('*******\nPuzzle1\n*******\n')

print('Test case\n---------\n')

res = update(read_input('input09a.txt'))

print(f'Checksum is {res}')

assert res == 1928

print('\nPuzzle case\n-----------\n')

res = update(read_input('input09.txt'))

print(f'Checksum is {res}')

assert res == 6262891638328

print('\n*******\nPuzzle2\n*******\n')

print('Test case\n---------\n')

res = reorder_and_check(read_input('input09a.txt'))

print(f'Checksum is {res}')

assert res == 2858

print('\nPuzzle case\n-----------\n')

res = reorder_and_check(read_input('input09.txt'))

print(f'Checksum is {res}')

assert res == 6287317016845
```
