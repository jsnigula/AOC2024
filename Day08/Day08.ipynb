```python
from collections import defaultdict
import math
```

```python
def read_input(infile):
    adict = {}
    pdict = defaultdict(list)
    ny, nx = 0, 0
    with open(infile, 'r') as inf:
        for y, line in enumerate(inf.readlines()):
            for x, c in enumerate(line.strip()):
                if c != '.':
                    pdict[c].append((y,x))
                    adict[(y,x)] = c
                nx = max(nx, x)
            ny = max(ny, y)
    return pdict, adict, ny, nx

def find_nodes(pdict, adict, ny, nx):
    nodes = []

    for a, pos in pdict.items():
        for i, (y1, x1) in enumerate(pos[:-1]):
            for y2, x2 in pos[i+1:]:

                if x2 > x1:
                    x_0 = x1
                    y_0 = y1
                    x_1 = x2
                    y_1 = y2
                else:
                    x_0 = x2
                    y_0 = y2
                    x_1 = x1
                    y_1 = y1
                    
                dx = x_1 - x_0
                dy = y_1 - y_0

                x_a = x_0 + 2*dx
                y_a = y_0 + 2*dy
                x_b = x_1 - 2*dx
                y_b = y_1 - 2*dy

                if y_a >= 0 and y_a <= ny and x_a >= 0 and x_a <= nx:
                    nodes.append((y_a, x_a))
                if y_b >= 0 and y_b <= ny and x_b >= 0 and x_b <= nx:
                    nodes.append((y_b, x_b))

    return len(set(nodes))

def find_nodes2(pdict, adict, ny, nx):
    nodes = []

    for a, pos in pdict.items():
        for i, (y1, x1) in enumerate(pos[:-1]):
            for y2, x2 in pos[i+1:]:

                if x2 > x1:
                    x_0 = x1
                    y_0 = y1
                    x_1 = x2
                    y_1 = y2
                else:
                    x_0 = x2
                    y_0 = y2
                    x_1 = x1
                    y_1 = y1
                    
                dx = x_1 - x_0
                dy = y_1 - y_0

                j = 0
                while (x_1 - j*dx) >= 0 or (x_0 + j*dx) <= nx:
                    x_a = x_0 + j*dx
                    y_a = y_0 + j*dy
                    x_b = x_1 - j*dx
                    y_b = y_1 - j*dy

                    if y_a >= 0 and y_a <= ny and x_a >= 0 and x_a <= nx:
                        nodes.append((y_a, x_a))
                    if y_b >= 0 and y_b <= ny and x_b >= 0 and x_b <= nx:
                        nodes.append((y_b, x_b))

                    j+=1

    return len(set(nodes))

find_nodes2(*read_input('input08.txt'))
```
