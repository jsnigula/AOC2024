```python
from collections import defaultdict
```

```python
def read_input(infile):
    pairs = []
    with open(infile, 'r') as inf:
        for line in inf.readlines():
            pairs.append(tuple(line.strip().split('-')))
    return pairs

def find_triplets_with_t(pairs):

    rings = defaultdict(set)

    for c1, c2 in pairs:
        rings[c1].add(c2)
        rings[c2].add(c1)

    triplets=set()
    for k, v in rings.items():
        for c in v:
            if c == k:
                continue
            for c2 in rings[c]:
                if c2==c or c2==k:
                    continue
                if k in rings[c2]:
                    triplets.add(tuple(sorted((k, c, c2))))

    n_t = 0
    for t in triplets:
        for n in t:
            if n[0] == 't':
                n_t += 1
                break
    return n_t

def find_largest_set(pairs):

    rings = defaultdict(set)

    for c1, c2 in pairs:
        rings[c1].add(c2)
        rings[c2].add(c1)

    triplets=set()
    for k, v in rings.items():
        for c in v:
            if c == k:
                continue
            for c2 in rings[c]:
                if c2==c or c2==k:
                    continue
                if k in rings[c2]:
                    triplets.add(tuple(sorted((k, c, c2))))

    sequences = []
    for t in triplets:
        t = tuple(t)
        t_set = set(t)
        for c in rings[t[0]]:
            found = True
            for tt in t_set:
                if c not in rings[tt]:
                    found = False
                    break
            if found:
                t_set.add(c)
        sequences.append(t_set)

    best = []
    for s in sequences:
        if len(s) > len(best):
            best = s
    return ','.join(sorted(list(best)))
```

```python
print('*******\nPuzzle1\n*******\n')

print('Test case\n---------\n')

res = find_triplets_with_t(read_input('input23a.txt'))

print(f'Number is {res}')

assert res == 7

print('\nPuzzle case\n-----------\n')

res = find_triplets_with_t(read_input('input23.txt'))

print(f'Number is {res}')

assert res == 1304

print('\n*******\nPuzzle2\n*******\n')

print('Test case\n---------\n')

res = find_largest_set(read_input('input23a.txt'))

print(f'Password is {res}')

assert res == 'co,de,ka,ta'

print('\nPuzzle case\n-----------\n')

res = find_largest_set(read_input('input23.txt'))

print(f'Password is {res}')

assert res == 'ao,es,fe,if,in,io,ky,qq,rd,rn,rv,vc,vl'
```
