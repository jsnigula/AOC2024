```python
from collections import defaultdict
```

```python
def read_input(infile):
    towels = defaultdict(list)
    patterns = []
    with open(infile, 'r') as inf:
        for t in [v.strip() for v in inf.readline().split(',')]:
            towels[t[0]].append((t, 1))
        inf.readline()
        for line in inf.readlines():
            patterns.append(line.strip())
    return patterns, towels
```

```python
def find_pattern(pattern, towels):

    patterns = [(pattern, [])]
    while len(patterns) != 0:
        p, m = patterns.pop()
        for t in towels[p[0]]:
            if p == t:
                return m + [t]
            l = len(t)
            if p[0:l] == t:
                # Match
                patterns.append((p[l:], m+[t]))

    return []                

def find_pattern_all(pattern, towels, combinations, n_comb):

    patterns = [(pattern, [])]
    matches = []

    # print('\nPattern', pattern)
    # print('Combinations', combinations)
    # print('N_Comb', n_comb)
    # print('lp', len(pattern))
    for idx in range(len(pattern)-1, -1, -1):

        p = pattern[idx:]

        # print('checking', p)
        if p in n_comb:
            # print('Known')
            continue

        if p[0] not in towels:
            continue
        lp = len(p)
        for t in towels[p[0]]:
            # print('testing', t)
            lt = len(t)
            if lt > lp:
                # print('Too long')
                continue
            if p[0:lt] == t:
                # print(p[0:lt], '==', t)
                l2 = lp - lt
                if l2 > 0:
                    # print('Searching for', p[lt:])
                    if p[lt:] in combinations[l2]:
                        # print('Adding', p)
                        combinations[lp].add(p)
                        n_comb[p] += n_comb[p[lt:]]
                else:
                    # print('Also adding', p)
                    combinations[lp].add(p)
                    n_comb[p] += 1
    return n_comb[p]
                    
def find_patterns(patterns, towels):

    n_found = 0
    for p in patterns:
        if len(find_pattern(p, towels)) != 0:
            n_found += 1
    return n_found

def find_patterns2(patterns, towels):

    n_found = 0

    # print(towels.values())
    combinations = defaultdict(set)
    n_comb = defaultdict(lambda: 0)
    for t in towels.values():
        for tt in t:
            # print('t is', tt)
            find_pattern_all(tt, towels, combinations, n_comb)
            # print('towels', towels)

    for p in patterns:
        n_found += find_pattern_all(p, towels, combinations, n_comb)
        

    return n_found
```

```python
print('*******\nPuzzle1\n*******\n')

print('Test case\n---------\n')

res = find_patterns(*read_input('input19a.txt'))

print(f'Number of designs is {res}')

assert res == 6

print('\nPuzzle case\n-----------\n')

res = find_patterns(*read_input('input19.txt'))

print(f'Number of designs is {res}')

assert res == 263

print('\n*******\nPuzzle2\n*******\n')

print('Test case\n---------\n')

res = find_patterns2(*read_input('input19a.txt'))

print(f'Number of designs is {res}')

assert res == 16

print('\nPuzzle case\n-----------\n')

res = find_patterns2(*read_input('input19.txt'))

print(f'Number of designs is {res}')

assert res == 723524534506343
```
