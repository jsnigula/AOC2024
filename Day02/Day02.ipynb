```python
def is_safe(report):

    direction = -1 if report[1] < report[0] else 1

    for i in range(1, len(report)):
        if 0 < direction*(report[i] - report[i-1]) < 4:
            continue
        return False
    return True
        
def is_safe_with_dampener(report):

    if is_safe(report):
        return True

    # Skip first element
    if is_safe(report[1:]):
        return True

    # Skip last element
    if is_safe(report[:-1]):
        return True

    for i in range(1, len(report)-1):
        if is_safe(report[0:i] + report[i+1:]):
            return True

    return False

def find_safe_reports(reports):
    n_safe = 0
    for r in reports:
        if is_safe(r):
            n_safe += 1

    return n_safe

def find_safe_reports_with_dampener(reports):
    n_safe = 0
    for r in reports:
        if is_safe(r):
            n_safe += 1
        else:
            if is_safe_with_dampener(r):
                n_safe += 1

    return n_safe
        
```

```python
def read_input(infile):
    reports = []
    with open(infile, 'r') as inf:
        for line in inf.readlines():
            reports.append([int(v) for v in line.strip().split()])

    return reports
```

```python
print('*******\nPuzzle1\n*******\n')

print('Test case\n---------\n')

res = find_safe_reports(read_input('input02a.txt'))

print(f'Number of safe reports is {res}')

assert res == 2

print('\nPuzzle case\n-----------\n')

res = find_safe_reports(read_input('input02.txt'))

print(f'Number of safe reports is {res}')

assert res == 383

print('\n*******\nPuzzle2\n*******\n')

print('Test case\n---------\n')

res = find_safe_reports_with_dampener(read_input('input02a.txt'))

print(f'Number of safe reports with dampener is {res}')

assert res == 4

print('\nPuzzle case\n-----------\n')

res = find_safe_reports_with_dampener(read_input('input02.txt'))

print(f'Number of safe reports with dampener is {res}')

assert res == 436
```
